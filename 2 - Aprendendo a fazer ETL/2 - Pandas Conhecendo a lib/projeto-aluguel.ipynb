## Análise Exploratoria dos dados

### 1 - Quais os valores médios de aluguel por tipo de imóvel?

```python
# Calcular a media dos valores da coluna valor 

dados["Valor"].mean()
```

```python
# Agrupar os dados de acordo com o tipo de imovel

dados.groupby("Tipo").mean(numeric_only=True)
```

```python
# Calcular a média dos valores agrupado por tipo ordenando do menor ao maior

dados.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
```

```python
# Criar um dataframe com a média dos valores agrupado por tipo
df_media_tipo = dados.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
```

```python
# Criar um grafico de barra horizontal com os dados do dataframe media_tipo
df_media_tipo.plot(kind="barh", figsize=(12, 8), color="orange")
```

### Remover os imoveis comerciais

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = [
    "Conjunto Comercial/Sala",
    "Prédio Inteiro",
    "Loja/Salão",
    "Galpão/Depósito/Armazém",
    "Casa Comercial",
    "Terreno Padrão",
    "Box/Garagem",
    "Loja Shopping/ Ct Comercial",
    "Chácara",
    "Loteamento/Condomínio",
    "Sítio",
    "Pousada/Chalé",
    "Hotel",
    "Indústria",
]
```

```python
dados.query('@imoveis_comerciais in Tipo')
```

```python
df_imoveis_n_comerciais = dados.query("@imoveis_comerciais not in Tipo")
```

```python
df_imoveis_n_comerciais.Tipo.unique()
```

```python
# Plotar um gráfico com a média dos valores dos tipo de imoveis não comerciais

df_media_tipo = df_imoveis_n_comerciais.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
df_media_tipo.plot(kind="barh", figsize=(10, 6), color="blue")
```

### 2 - Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
df_imoveis_n_comerciais.Tipo.value_counts(normalize=True)
```

```python
df_imoveis_n_comerciais.Tipo.value_counts(normalize=True).to_frame().sort_values("Tipo")
```

```python
df_percentual_tipo = (
    df_imoveis_n_comerciais.Tipo.value_counts(normalize=True)
    .to_frame()
    .sort_values("Tipo")
)

df_percentual_tipo.plot(kind="bar", figsize=(10,8), color="blue", xlabel="Tipo", ylabel="Percentual")
```

```python
# selecionar apenas  os imoveis do tipo apartamento

df_apartamento = df_imoveis_n_comerciais.query('Tipo == "Apartamento"')
```
